# 8

```python
df["pub_day"] = df["publication_date"].dt.day
df["pub_month"] = df["publication_date"].dt.month
df["pub_year"] = df["publication_date"].dt.year

df.head()
```

# 9

```python
dates = df[df["title"].str.contains("Harry Potter")]["publication_date"]
first_date = dates.min()
last_date = dates.max()

delta = last_date - first_date

print(delta.days)
```

# 13

```python
import sqlite3


conn = sqlite3.connect("../data/books.db")
```

Ce code crée une connexion vers une base de données SQLite qui sera sauvegardée à l'emplacement `../data/books.db`.

# 14

```python
selected = df[["bookID", "title", "language_code", "publisher", "pub_year"]]

selected.head()
```

# 15

```python
selected.to_sql(name="Book", con=conn)
```

```python
author_df = df[["bookID", "authors"]].copy()
author_df["authors"] = author_df["authors"].str.split("/")
author_df = author_df.explode("authors")
display(author_df.head(10))

author_df.to_sql(name="Book_authors", con=conn)
```
